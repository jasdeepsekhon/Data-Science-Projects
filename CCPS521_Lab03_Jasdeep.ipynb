```python
import sklearn  # machine learning library for Python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
iris = datasets.load_iris()  # load and return iris dataset
print(iris.data.shape)  # print shape of data to confirm if it is loaded
data1 = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                     columns=iris['feature_names'] + ['target'])
print(data1)
```

```python
Flower = []

for i in range(len(iris['target'])):
    if iris['target'][i] == 0:
        Flower.append('setosa')
    elif iris['target'][i] == 1:
        Flower.append('versicolor')
    else:
        Flower.append('virginica')

iris['Flower'] = Flower

data1 = pd.DataFrame(data=np.c_[iris['data'], iris['target'], iris['Flower']],
                     columns=iris['feature_names'] + ['target'] + ['Flower'])

print(data1)
```

```python
data1["petal length (cm)"] = pd.to_numeric(data1["petal length (cm)"])
data1["petal width (cm)"] = pd.to_numeric(data1["petal width (cm)"])
data1["sepal length (cm)"] = pd.to_numeric(data1["sepal length (cm)"])
data1["sepal width (cm)"] = pd.to_numeric(data1["sepal width (cm)"])
```

```python
corr = data1.corr()
sns.heatmap(corr,annot=True, fmt='.3f',xticklabels=corr.columns.values,yticklabels=corr.columns.values)
plt.show()
```

```python
linear_regressor = LinearRegression()
X = data1.iloc[:, 2].values.reshape(-1, 1)
Y = data1.iloc[:, 3].values.reshape(-1, 1)
linear_regressor.fit(X, Y)
Y_pred = linear_regressor.predict(X)
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```
