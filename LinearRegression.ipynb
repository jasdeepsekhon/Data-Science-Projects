```python
import pandas as pd

import sklearn  # machine learning library for Python
import numpy as np
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn import linear_model 
import statsmodels.api as sm
# load and read dataset mtcars
```

```python
dataset = pd.read_fwf(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data')
```

```python
cars_df = pd.DataFrame(dataset)
cars_df.columns = ["mpg", "cyl", "disp",
                   "hp", "wt", "acc", "year", "org", "name"]
```

```python
clean_cars_df = cars_df.dropna()
clean_cars_df = clean_cars_df.dropna(how='all')
clean_cars_df = clean_cars_df.drop(columns=["name"])
```

```python
clean_cars_df["mpg"] = pd.to_numeric(clean_cars_df["mpg"])
clean_cars_df["cyl"] = pd.to_numeric(clean_cars_df["cyl"])
clean_cars_df["disp"] = pd.to_numeric(clean_cars_df["disp"])
clean_cars_df = clean_cars_df[clean_cars_df.hp != "?"]
clean_cars_df["hp"] = pd.to_numeric(clean_cars_df["hp"])
clean_cars_df["wt"] = pd.to_numeric(clean_cars_df["wt"])
clean_cars_df["acc"] = pd.to_numeric(clean_cars_df["acc"])
clean_cars_df["year"] = pd.to_numeric(clean_cars_df["year"])
clean_cars_df["org"] = pd.to_numeric(clean_cars_df["org"])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 10))
start_index = 2
for i, ax in enumerate(axs.flat):
    ax.boxplot(clean_cars_df.iloc[:, i+start_index])
    ax.set_title(
        clean_cars_df.columns[i+start_index], fontsize=20, fontweight='bold')
    ax.tick_params(axis='y', labelsize=14)
plt.show()
```

```python
x = clean_cars_df[['disp', 'hp', 'wt']]
y = clean_cars_df['mpg']

# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
print('r squared: ', regr.score(x,y))
# with statsmodels
x = sm.add_constant(x)  # adding a constant

model = sm.OLS(y, x).fit()
predictions = model.predict(x)

print_model = model.summary()
print(print_model)
```
